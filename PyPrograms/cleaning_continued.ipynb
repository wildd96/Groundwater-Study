```python
for files in os.listdir(piemonte_filepath):
    if files.endswith('.xlsx'):
        d = pd.read_excel(piemonte_filepath + '\\' + files)
        d = d.rename(columns={'Denominazione':'codice', 'Data':'data', 'LIVELLO PIEZOMETRICO - m':'val', 'SOGGIACENZA - m':'val-inv'})
        d = d[['codice', 'data', 'val', 'val-inv']]
        d['data'] = pd.to_datetime(d['data']);
        d = d.reset_index(drop=True)
        d = d.sort_values('data', ascending=True)
        d.to_excel(piemonte_filepath + '\\' + files)
        combined_table = combined_table.append(d)
combined_table = combined_table.reset_index(drop = True)
combined_table.to_excel(piemonte_filepath + '\\combined-tables\\COMBINED_PIEMONTE_TABLE.xlsx')
```

```python
combined_table = pd.DataFrame()

for files in os.listdir(trento_filepath):
    if files.endswith('.xlsx'):
        d = pd.read_excel(trento_filepath + '\\' + files)
        d = d.rename(columns={0:'data', 2:'val'})
        d['codice'] = files[:7]
        d = d[['codice', 'data', 'val']]
        d['data'] = pd.to_datetime(d['data'])
        d = d.sort_values('data', ascending=True)
        d = d.reset_index(drop=True)
        year_min = str(d['data'].min().year)
        year_max = str(d['data'].max().year)
        d.to_excel(trento_filepath + '\\' + files[:7] + '_' + year_min + '_' + year_max + '.xlsx')
        combined_table = combined_table.append(d)
combined_table = combined_table.reset_index(drop = True)
combined_table.to_excel(trento_filepath + '\\combined-tables\\COMBINED_TRENTO_TABLE.xlsx')
```
