```python
import numpy as np
import pandas as pd
import os, sys
import codecs
import utm
from pyproj import Proj, transform
from scipy.io import savemat
```

```python
cwd = os.getcwd()
```

```python
directory_name = "/FormattedEmilia"
original_file_type = ".xlsx"
whole_table = pd.DataFrame()
#change last part depending on imported file type
for files in os.listdir(cwd + directory_name):
    if files.endswith(original_file_type):
        
        file_path = cwd + directory_name + "/" + files
        
        df = pd.read_excel(file_path)
    
        arr = []
        for i in range(len(df['Piezometria (m)'])):
            if type(df['Piezometria (m)'].values[i]) == str:
                arr = np.append(arr, i)

        for j in arr:
            df = df[df.index != j]
            
        whole_table = whole_table.append(df)
        
        xls = pd.ExcelFile(cwd + directory_name + "/Anagrafica/dati_monitoraggio_acque_sotterranee_2010-2012.xlsx")
        
df = pd.read_excel(xls, 'Piezometria_Soggiacenza')

df = df[['Codice', 'Data', 'Piezometria (m)', 'Soggiacenza (m)']]

whole_table = whole_table[['Codice', 'Data', 'Piezometria (m)', 'Soggiacenza (m)']]
        
whole_table = whole_table.append(df)

whole_table
```

```python
date_arr = []
for i in whole_table['Data']:
    date_arr = np.append(date_arr, (str(i)[:10]))
    
whole_table['Data'] = date_arr
```

```python
for i in df['Codice_RER']:
    hold = df
    inProj = Proj(init='epsg:5659');
    outProj = Proj(init='epsg:4326');
    x1,y1 = hold['X_UTM-ED50'], hold['Y_UTM-ED50'];
    x2,y2 = transform(inProj,outProj,x1,y1);
    hold['Longitude'] = x2;
    hold['Latitude'] = y2;
```

```python
if not (os.path.exists(cwd + "/processed_data/"+ directory_name + "_processed")): 
            os.makedirs(cwd + "/processed_data/"+ directory_name + "_processed")
        
for i in whole_table['Codice'].unique():
    hold = whole_table[whole_table['Codice'] == i]
    max_year = str(hold[hold['Codice'] == i]['Data'].max())[:4]
    min_year = str(hold[hold['Codice'] == i]['Data'].min())[:4]
    
    hold = hold.sort_values(by = ['Data'], ascending = False)

    if len(hold) > 10:
        hold.to_excel(cwd + "/processed_data" + directory_name + "_processed/" + i + "_" + min_year + "_" + max_year + '.xlsx',
                      index = False)
        if not (os.path.exists(cwd + "/processed_data" + directory_name + "_processed/mat_files/")): 
                os.makedirs(cwd + "/processed_data" + directory_name + "_processed/mat_files/")
        savemat(cwd + "/processed_data" + directory_name + "_processed/mat_files/" + str(i) + "_" +
                min_year + "_" + 
                max_year + ".mat", hold)
```
