```python
import numpy as np
import pandas as pd
import os, sys
import codecs
from scipy.io import savemat
```

```python
cwd = os.getcwd()
```

```python
file_path = "/Users/andrew/Desktop/girotto_Lab/GW/ValdAosta_misure quantitative.xlsx"
```

```python
xls = pd.ExcelFile(file_path)
df1 = pd.read_excel(xls, 'punti', header = None).iloc[:,:4]
df2 = pd.read_excel(xls, 'misure', header = None)

df1 = df1.rename({0: "station_id"}, axis = 1)
df1
```

```python
df = df2.iloc[:,1:20]
dates = df2.iloc[:, 0]
df
```

```python
df2[1][1]
```

```python
x = df.shape[1] + 1
y = df.shape[0]
y
```

```python
directory_name = 'ValdAosta_misure quantitative'

whole_table = pd.DataFrame()

for i in np.arange(1,x):
    station_id = []
    quota = []
    date = []
    for j in np.arange(1,y):
        
        station_id = np.append(station_id, df2[i][0])
        quota = np.append(quota, df2[i][j])
        date = np.append(date, df2[0][j])
        
        table = {
            "data": date,
            "station_id":station_id,
            "quota":quota
        }
        
    table = pd.DataFrame(data = table)
    table = pd.merge(table, df1, on = 'station_id')
    table = table.rename({1: 'Manuale/Datalogger', 2: 'Longitude', 3: 'Latitude'}, axis = 'columns')
    table = table[table['quota'] != '#N/D']
    table = table[table['quota'] != None]
    table = table[table['quota'] != 'nan']

    whole_table = whole_table.append(table)
    
    table = table[['data', 'station_id', 'quota']]
        
    if not (os.path.exists(cwd + "/processed_data/"+ directory_name + "_processed")): 
            os.makedirs(cwd + "/processed_data/"+ directory_name + "_processed")
    
    table.to_excel( "/Users/andrew/Desktop/girotto_Lab/GW/processed_data/" + directory_name + "_processed/" 
                      + max(table['station_id']) + "_" + table['data'].min()[-4:] + "_" + table['data'].max()[-4:] + ".xlsx", index = False)
    
    if not (os.path.exists("/Users/andrew/Desktop/girotto_Lab/GW/processed_data/" + directory_name + "_processed/mat_files/")): 
            os.makedirs(cwd + "/processed_data/" + directory_name + "_processed/mat_files/")
            
    # savemat("/Users/andrew/Desktop/girotto_Lab/GW/processed_data/mat_files/" + directory_name + "_processed/" 
    #                   + max(table['station_id']) + "_" + table['data'].min()[-4:] + "_" + table['data'].max()[-4:], table)
```
